```python
import pandas as pd
start_date = str('2016-01-01') 
end_date = str('2016-12-31')
company = Share('AAPL')
d = []
for p in range(len(company.get_historical(start_date, end_date))-1, -1, -1):
    tmp = company.get_historical(start_date, end_date)[p]
    d.append({'ADate': tmp.get('Date'),\
    'Open': tmp.get('Open'), \
    'Low': tmp.get('Low'), \
    'High': tmp.get('High'), \
    'Close': tmp.get('Close'), \
    'Volume': tmp.get('Volume')})

d = pd.DataFrame(d)
d.head()
APPLE_2016_Stock = d
```

```python
#Apple Price 2015 last date = 105.26
import pandas as pd
import numpy as np
start_date = str('2016-01-01') 
end_date = str('2016-12-31')
company = Share('AAPL')

label = []
length = len(company.get_historical(start_date, end_date))
threshold = 2
for p in range(length-1, -1, -1):
    curr_price = float(company.get_historical(start_date, end_date)[p].get('Close'))
    if (p + 1) <= (length - 1):
        priv_price = float(company.get_historical(start_date, end_date)[p + 1].get('Close'))
    else:
        priv_price = 105.26
    
    if ((curr_price - priv_price)/priv_price)*100 > threshold:
        indicator = 1
    elif (((curr_price - priv_price) >= 0) and (((curr_price - priv_price)/priv_price)*100 < threshold)) or \
     (((curr_price - priv_price) < 0) and (((curr_price - priv_price)/priv_price)*100 > -1*threshold)):
        indicator = 0
    else:
        indicator = -1
    label.append({'ADate': company.get_historical(start_date, end_date)[p].get('Date'), \
    'Performance': indicator})
    

label = pd.DataFrame(label)
label.head()
APPLE_2016_Performance = label
print (APPLE_2016_Performance)
```

```python
APPLE_2016_Performance.to_csv(r'/home/cloudera/APPLE_2016_Performance.txt', header=None, index=None, sep=' ', mode='a')
!cat APPLE_2016_Performance.txt
```
